<b> Importing Libraries <b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# <b> Loading DataSet <b>

```python
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
df.head()
```

# <b> Data Cleaning <b> 

<b> Checking Missing Values <b>

```python
df.isnull().sum()
```

<b> Remove Missing Values <b>

```python
df.dropna(inplace=True)
```

<b> Removing Duplicates <b>

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

# <b> Data Analysis + Visualizations <b>

<b> 1. Sales Trend (Line Chart) <b>

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.month
monthly_sales = df.groupby('Month')['Total_Sales'].sum()
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

# <b> Top Products Analysis <b>

```python
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
top_products.head(10).plot(kind='bar')
plt.title("Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

# <b> Category Analysis <b>

```python
category_sales = df.groupby('Category')['Total_Sales'].sum()
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Distribution")
plt.show()
```

# <b> City-wise Sales Analysis <b> 

```python
city_sales = df.groupby('City')['Total_Sales'].sum()
city_sales.plot(kind='bar')
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

 <b> Key Insights

- Sales vary across different months.
- Some products contribute more to total sales.
- Category distribution helps identify high-performing categories.
- Certain cities generate higher sales than others.
- Visual analysis helps understand business performance better. <b>
